```python
import yfinance as yf
import datetime
import pandas as pd
from datetime import datetime
from bokeh.resources import INLINE
from bokeh.embed import components

from bokeh.plotting import figure, show, save
from bokeh.io import output_notebook
from bokeh.palettes import Dark2_5 as palette
import itertools

colors = itertools.cycle(palette) #there are some new imports, make sure that app.py has all of them

output_notebook()
```

```python
ticker = [] #set list of ticker
```

```python
data = yf.download(tickers = ticker, period='ytd')
data = data.reset_index()
p = figure(title="Year to Date Performance", x_axis_label = "Date", y_axis_label="Stock Price", x_axis_type="datetime", plot_width=800, plot_height=300)
for x in ticker:
    p.line(y = data['Adj Close'][x], x = data["Date"], color=next(colors),  legend_label=x)
p.legend.location = "top_left"
p.legend.click_policy="hide"
plot = show(p)
```

```python
js_resources = INLINE.render_js()
css_resources = INLINE.render_css()
script, div = components(p)

## this is for the flask, look at the code that I wrote
```
