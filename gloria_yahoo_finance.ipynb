```python
import yfinance as yf
import datetime
import pandas as pd
from datetime import datetime

from bokeh.plotting import figure, show, save
from bokeh.io import output_notebook

output_notebook()
```

```python
ticker = 'MSFT'
obj = yf.Ticker(ticker)
country = obj.info['country']
website = obj.info['website']
name = obj.info['longName']
bus = obj.info['longBusinessSummary']
Vol = obj.info['averageVolume10days']
beta = obj.info['beta']
TPE = obj.info['trailingPE']
PEG = obj.info['pegRatio']
ADR = obj.info['trailingAnnualDividendRate']
BV = obj.info['bookValue']
bid = obj.info['bid']
ask = obj.info['ask']
vol = obj.info['volume']
vol10 = obj.info['averageDailyVolume10Day']
low = obj.info['regularMarketDayLow']
high = obj.info['regularMarketDayHigh']
low2 = obj.info['dayLow']
high2 = obj.info['dayHigh']
mclose = obj.info['previousClose']
mopen = obj.info['regularMarketOpen']
print(bid, ask, vol, vol10, low, high, low2, high2, mclose, mopen)
```

```python
ticker = input('Please enter a ticker ')
date   = input('Please enter a time period (5d|1mo|3mo|6mo|1y|2y|5y|10y|ytd|max) ')
ticker = ticker.lower()
data = yf.download(tickers = ticker, period=date)
data = data.reset_index()
title = ''.join(("Stock chart of ", ticker.upper()))
p = figure(title=title, x_axis_label = "Date", y_axis_label="Stock Price", x_axis_type="datetime", plot_width=800, plot_height=300)
p.line(y = data["Open"], x = data["Date"], color="green")
show(p)
```
